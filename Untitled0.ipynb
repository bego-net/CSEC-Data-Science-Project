 2,Given the list [10, 15, 20, 25, 30], write a function that returns the square of only even numbers.

```python
num_list = [10, 15, 20, 25, 30]
even_square = []

for i in num_list:
    if i % 2 == 0:
        even_square.append(i * i)

print(even_square)
```

7.  What does the .groupby() function do in pandas? Show an example using dummy data.

The .groupby() function in pandas is used to group the data by some category (like gender,class,department etc.) so you can do calculations on each group like getting the average, sum, or count.

```python
import pandas as pd

data = {
    "Class": ["A", "A", "B", "B", "C", "C"],
    "Student": ["Bego", "Liya", "Robel", "Hana", "Aman", "Sami"],
    "Score": [85, 90, 78, 88, 92, 80]
}

df = pd.DataFrame(data)
# Group by class and find the average score for each class
avg_scores = df.groupby("Class")["Score"].mean()

print(avg_scores)
```

10, . What does it mean if a column has a high standard deviation?

  Create a list or column that has a low and a high std. dev, and compare.  

It tells us how spread out the values are in a column or list.
high standard deviation: values are far apart (more spread)
low standard deviation: values are close together (less spread)

```python
import pandas as pd

low_std_list = [100, 101, 99, 100, 102]
low_std = pd.Series(low_std_list)

high_std_list = [10, 200, 50, 150, 100]
high_std = pd.Series(high_std_list)

print("Low std dev list:", low_std_list)
print("   → Std Dev:", round(low_std.std(), 2))

print()

print("High std dev list:", high_std_list)
print("   → Std Dev:", round(high_std.std(), 2))
```
